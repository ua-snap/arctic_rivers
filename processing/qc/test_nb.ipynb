```python
import xarray as xr
from pathlib import Path

# define the data directory
data_dir = Path('/import/beegfs/CMIP6/arctic-cmip6/Arctic_Rivers_Data')

# list all .nc files with *Q.nc or *WT.nc extensions and sort
all_files = list(data_dir.rglob('*[Q|WT].nc'))
all_files.sort()

# print the range of the time dimension for each file
for file in all_files:
    ds = xr.open_dataset(file, decode_cf=True)
    time_range = (str(ds['time'].min().values)[:10], str(ds['time'].max().values)[:10])
    print(f"File: {file.name}, Time Range: {time_range}")
```
