## 3. Simplify Segments (Arctic Rivers)

Simplifies segment geometries for web delivery and writes two output shapefiles:
- `export/arctic_rivers_segments_stats_simplified.shp` — simplified geometry with stats
- `export/arctic_rivers_segments_stats_simplified_subset.shp` — simplified geometry, IDs only (no stats)

**Input:** `export/arctic_rivers_segments_stats.shp` (from `add_stats_to_shp.ipynb`)  
**CRS:** EPSG:3338 (Alaska Albers, metric) — tolerance=50m matches CONUS approach

```python
import geopandas as gpd
from pathlib import Path
```

```python
INPUT_PATH      = "export/arctic_rivers_segments_stats.shp"
OUTPUT_PATH     = "export/arctic_rivers_segments_stats_simplified.shp"
OUTPUT_SUBSET   = "export/arctic_rivers_segments_stats_simplified_subset.shp"
SIMPLIFY_TOL    = 50  # metres (EPSG:3338 is metric)
```

```python
gdf = gpd.read_file(INPUT_PATH)
print(f"Input: {len(gdf)} segments, CRS: {gdf.crs}")
```

```python
gdf_simplified = gdf.copy()
gdf_simplified["geometry"] = gdf.simplify(SIMPLIFY_TOL, preserve_topology=True)

print(f"Simplified with tolerance={SIMPLIFY_TOL}m")
print(f"Output: {len(gdf_simplified)} segments")
```

```python
gdf_simplified.to_file(OUTPUT_PATH)
print(f"Wrote {OUTPUT_PATH}")

# Subset: geometry + COMID only (no stat columns)
gdf_subset = gdf_simplified[["COMID", "geometry"]]
gdf_subset.to_file(OUTPUT_SUBSET)
print(f"Wrote {OUTPUT_SUBSET}")
```
