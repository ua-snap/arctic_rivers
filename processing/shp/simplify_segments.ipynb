## 3. Simplify Segments (Arctic Rivers)

Simplifies segment geometries for web delivery and writes two output shapefiles:
- `export/arctic_rivers_segments_stats_simplified.shp` — simplified geometry with stats
- `export/arctic_rivers_segments_stats_simplified_subset.shp` — simplified geometry, IDs only (no stats)

**Input:** `export/arctic_rivers_segments_stats.shp` (from `add_stats_to_shp.ipynb`)  
**CRS:** EPSG:3338 (Alaska Albers, metric) — tolerance=50m matches CONUS approach

```python
import geopandas as gpd
from pathlib import Path
```

```python
STATS_INPUT_PATH      = "segments/arctic_rivers_segments_stats.shp"
STATS_OUTPUT_PATH     = "export/arctic_rivers_segments_stats_simplified.shp"

# Simplify the version of the segment file that includes HUC and outlet information.
WITH_HUC_INPUT_PATH   = "segments/arctic_rivers_segments_joined_3338.shp"
WITH_HUC_OUTPUT_PATH  = "export/arctic_rivers_segments_joined_3338_simplified.shp"

SIMPLIFY_TOL    = 50  # metres (EPSG:3338 is metric)
```

```python
stats_gdf = gpd.read_file(STATS_INPUT_PATH)
print(f"Stats file input: {len(stats_gdf)} segments, CRS: {stats_gdf.crs}")

with_huc_gdf = gpd.read_file(WITH_HUC_INPUT_PATH)
print(f"With HUC file input: {len(with_huc_gdf)} segments, CRS: {with_huc_gdf.crs}")
```

```python
stats_gdf_simplified = stats_gdf.copy()
stats_gdf_simplified["geometry"] = stats_gdf.simplify(SIMPLIFY_TOL, preserve_topology=True)
print(f"Stats file simplified with tolerance={SIMPLIFY_TOL}m")
print(f"Output: {len(stats_gdf_simplified)} segments")


with_huc_gdf_simplified = with_huc_gdf.copy()
with_huc_gdf_simplified["geometry"] = with_huc_gdf.simplify(SIMPLIFY_TOL, preserve_topology=True)
print(f"With HUC file simplified with tolerance={SIMPLIFY_TOL}m")
print(f"Output: {len(with_huc_gdf_simplified)} segments")
```

```python
# Make an export directory if one doesn't exist
!mkdir -p export

stats_gdf_simplified.to_file(STATS_OUTPUT_PATH)
print(f"Wrote {STATS_OUTPUT_PATH}")

with_huc_gdf_simplified.to_file(WITH_HUC_OUTPUT_PATH)
print(f"Wrote {WITH_HUC_OUTPUT_PATH}")
```
